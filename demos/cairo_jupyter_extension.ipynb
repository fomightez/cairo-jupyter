```python
# Using display with cairo surfaces
import cairocffi as cairo

surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, 300, 300)
    
cr = cairo.Context(surface)

for i in range(16):
    cr.save()
    cr.translate(150, 150)
    cr.set_source_rgba(0.7, 0.9, 0.0, 0.25)
    cr.rotate(i * 0.1)
    cr.rectangle(-50, -50, 100, 100)
    cr.stroke()
    cr.restore()

display(surface)
```

```python
# Using display with cairo context to display the current state
import cairocffi as cairo

surface = cairo.ImageSurface(cairo.FORMAT_ARGB32, 50, 50)
    
cr = cairo.Context(surface)

for i in range(8):
    cr.save()
    cr.translate(30, 30)
    cr.set_source_rgba(0.7, 0.9, 0.0, 0.25)
    cr.rotate(i * 0.1)
    cr.rectangle(-10, -10, 20, 20)
    cr.stroke()
    cr.restore()

    display(cr)
```
